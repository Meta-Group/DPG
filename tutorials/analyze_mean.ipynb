```python
def combine(directory_path):
    all_files = os.listdir(directory_path)
    
    cluster_files = [f for f in all_files if 'clusters_count.csv' in f]
    
    dataframes = []
    indexes = []
    for file in cluster_files:
        file_path = os.path.join(directory_path, file)
        df = pd.read_csv(file_path)
        df = df.sort_values(by=df.columns[0], ascending=True)
        indexes.append(df.iloc[:, 0])
        df = df.drop(columns=["Unnamed: 0"], errors="ignore")
        df = df.reset_index(drop=True)
        dataframes.append(df)
    combined_df = pd.concat(dataframes, axis=1)
    combined_df.insert(0, 'Index', indexes[0].sort_values(ascending=True).reset_index(drop=True))
    
    return combined_df
```

```python
df = combine(path)
```

```python
class_columns = ['Class 0', 'Class 1', 'Class 2', 'Ambiguous']
dataframes = []
for i in class_columns:
    df_temp = round(df[i].mean(axis=1), 2)
    df_temp = df_temp.reset_index(drop=True)
    dataframes.append(df_temp)
mean_df = pd.concat(dataframes, axis=1)
mean_df.columns = class_columns
mean_df.insert(0, 'Index', df['Index'].sort_values(ascending=True).reset_index(drop=True))
mean_df
```
